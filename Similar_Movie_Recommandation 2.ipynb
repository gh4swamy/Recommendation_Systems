```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'

import pandas as pd, numpy as np
movies= pd.read_csv('data/movies.csv')
ratings = pd.read_csv('data/ratings.csv')
```

```python
movies.head()
movies.shape
ratings.head()
ratings.shape
```

```python
movie_ratings = pd.merge(movies, ratings, on='movieId')
movie_ratings.head()
movie_ratings.shape
```

```python
movie_ratings_stats = movie_ratings.groupby('title').agg({'rating':[np.size, np.mean]})
movie_ratings_stats.head(10)
```

```python
movie_ratings_stats.sort_values(by=[('rating', 'size')]).head(10)
```

```python
movie_ratings_stats.sort_values(by=[('rating', 'mean')], ascending=False).head(10)
```

```python
popular_movies = movie_ratings_stats['rating']['size']>80
popular_movies_df = movie_ratings_stats[popular_movies]
popular_movies_df.shape
popular_movies_df.head()
```

```python
rating_matrix=movie_ratings.pivot_table(index='userId', columns='title', values='rating')
rating_matrix.head()
```

```python
ratings_of_toy_story = rating_matrix['Toy Story (1995)']
ratings_of_toy_story.head(10)
```

```python
similar_to_toy_story = rating_matrix.corrwith(ratings_of_toy_story)
similar_to_toy_story = similar_to_toy_story.dropna()
```

```python
similar_to_toy_story_df = pd.DataFrame(similar_to_toy_story, columns=['similarity_index'])
similar_to_toy_story_df.head()
```

```python
final_result = popular_movies_df.join(similar_to_toy_story_df)
final_result
```
