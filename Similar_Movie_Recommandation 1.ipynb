```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'

import pandas as pd, numpy as np
movies= pd.read_csv('data/movies.csv')
ratings = pd.read_csv('data/ratings.csv')
```

```python
movies.shape
ratings.shape
movies.head(2)
```

```python
df = pd.merge(movies, ratings, how='inner', on = 'movieId')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
movie_ratings = df.pivot_table(index='userId', columns='title', values='rating')
movie_ratings.head()
```

```python
starwarratings = movie_ratings['Star Wars: Episode I - The Phantom Menace (1999)']
starwarratings.head(20)
```

```python
similar_to_starwars = movie_ratings.corrwith(starwarratings)

similar_to_starwars = similar_to_starwars.dropna()
similar_to_starwars_df=pd.DataFrame(similar_to_starwars)
similar_to_starwars_df.head()
```

```python
similar_to_starwars_df.sort_values(by=0, ascending=False).head(20)
```

```python
movie_ratings_size_factor = df.groupby('title').agg({'rating':[np.size, np.mean]})
movie_ratings_size_factor.head(20)
```

```python
popular_movies = movie_ratings_size_factor['rating']['size']>100
movie_ratings_size_factor[popular_movies].head()
```

```python
movie_ratings_size_factor[popular_movies].sort_values(by=[('rating', 'mean')], ascending=False)[:20]
```

```python
similar_to_starwars.head()
```

```python
final_result = movie_ratings_size_factor[popular_movies].join(pd.DataFrame(similar_to_starwars, columns=['similarity_index']))
final_result.head()
```

```python
final_result.sort_values(by='similarity_index', ascending=False)
```
